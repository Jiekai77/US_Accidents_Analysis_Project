```python
#clean data 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('US_Accidents_Dec20.csv')
```

```python
data.head()
```

```python
# check missing values of each column for training dataset
data.isnull()
data.isnull().sum()
```

```python
len(data)
```

```python
# if the number of missing values in the column is greater than half number of the rows, then we could delete this column
data.drop(['End_Lat'], axis=1, inplace=True)
data.drop(['End_Lng'], axis=1, inplace=True)
data.drop(['Number'], axis=1, inplace=True)
```

```python
# drop insignificant columns 
#data.drop(['Street', 'Airport_Code', 'Source'], axis=1, inplace=True)# drop insignificant columns 
data.drop(['Timezone', 'Civil_Twilight', 'Nautical_Twilight','Astronomical_Twilight'],axis=1, inplace=True)
```

```python
#data.drop(['Weather_Timestamp'],axis=1, inplace=True)
data.drop(['Side'],axis=1, inplace=True)
```

```python
data
```

```python
data['TMC'].fillna(data['TMC'].mean(),inplace=True)
```

```python
data['City'].fillna(data['City'].mode(),inplace=True)
```

```python
data['Zipcode'].fillna(data['Zipcode'].mode(),inplace=True)
```

```python
for column in ['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)','Pressure(in)', 'Visibility(mi)','Wind_Speed(mph)','Precipitation(in)']:
    data[column].fillna(data[column].mean(), inplace=True)
```

```python
for column in ['Weather_Condition', 'Wind_Direction', 'Sunrise_Sunset']:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

```python
result= data
result.to_csv('Cleaned data from Sophia')
```
